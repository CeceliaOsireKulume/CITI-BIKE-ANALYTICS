```python
# Create a subset containing only numerical variables
df_numerical = df.drop(columns=['trip_id', 'weekday', 'start_time', 'end_time', 'start_station_name', 'end_station_name', 'gender_label', 'start_month', 'end_station_latitude', 'end_station_longitude', 'subscriber', 'start_station_latitude', 'start_station_longitude'])
```

```python
df_numerical.head()
```

```python
### Changing 'NAN' values to 'column means' for analysis because it cuts a cross two variables that are key for analysis

df_numerical_new = df_numerical.fillna(df_numerical.mean())
df_numerical_new.head()
```

```python
df_numerical_new.loc[(df_numerical_new['Age']>= 10) & (df_numerical_new['Age']< 17), 'Age Category']='Children'
```

```python
df_numerical_new.loc[(df_numerical_new['Age']>= 17) & (df_numerical_new['Age']< 24), 'Age Category']='Youth'
```

```python
df_numerical_new.loc[(df_numerical_new['Age']>= 24) & (df_numerical_new['Age']< 64), 'Age Category']='Adults'
```

```python
df_numerical_new.loc[(df_numerical_new['Age']>= 64) & (df_numerical_new['Age']< 120), 'Age Category']='Seniors'
```

```python
df_numerical_new['Age Category'].value_counts()
```

```python
#Change Hours to PM/AM
```

```python
# Converting the time variable in the 'start_hour' and 'end_hour' to AM $ PM

def convert_ampm(hour):
    if hour == 0:
        return "12 AM"
    elif hour < 12:
        return f"{hour} AM"
    elif hour == 12:
        return "12 PM"
    else:
        return f"{hour - 12} PM"
```

```python
df_numerical_new['Start_time'] = df_numerical_new['start_hour'].apply(convert_ampm)
```

```python
df_numerical_new['End_time'] = df_numerical_new['end_hour'].apply(convert_ampm)
```
