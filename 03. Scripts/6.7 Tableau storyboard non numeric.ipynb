```python
# Converting the time variable in the 'start_hour' and 'end_hour' to AM $ PM

def convert_ampm(hour):
    if hour == 0:
        return "12 AM"
    elif hour < 12:
        return f"{hour} AM"
    elif hour == 12:
        return "12 PM"
    else:
        return f"{hour - 12} PM"
```

```python
df['Start_time'] = df['start_hour'].apply(convert_ampm)
```

```python
df['End_time'] = df['end_hour'].apply(convert_ampm)
```
